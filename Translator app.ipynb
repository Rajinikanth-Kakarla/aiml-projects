# Task 3: Translator App:
## You have to create a translator AI which translate between any two languages. Its upon you to choose language.



```python
from tkinter import *
from tkinter import ttk, messagebox
import googletrans
from googletrans import Translator
```

```python
root = Tk()
root.title("Radhe Translator")
root.geometry("347x575")
root.resizable(False, False)
```

```python
image_icon = PhotoImage(file = "radhe.png")
root.iconphoto(False,image_icon)
```

```python
root.config(bg="#D3D3D3")
```

```python
def label_change():
    c = combo1.get()
    c1 = combo2.get()
    label1.configure(text = c)
    label2.configure(text = c1)
    root.after(100,label_change)
```

```python
def translate_now():
    text_ = text1.get(1.0,END)
    t1 = Translator()
    trans_text = t1.translate(text_,src=combo1.get(),dest = combo2.get())
    trans_text = trans_text.text
    text2.delete(1.0,END)
    text2.insert(END,trans_text)
```

```python
language = googletrans.LANGUAGES
languageV = list(language.values())
lang1 = language.keys()
```

```python
combo1 = ttk.Combobox(root,values = languageV,font = "Roboto 14", state = "r")
combo1.place(x = 53, y = 15)
combo1.set("ENGLISH")
```

```python
label1 = Label(root,text = "ENGLISH",font = "segoe 30 bold", bg = "white",width = 12, bd=5, relief = GROOVE)
label1.place(x = 25, y = 50)
```

```python
f = Frame(root, bg = "Black",bd = 1)
f.place(x = 10,y = 118,width = 330, height = 125)
```

```python
text1 = Text(f, font = "Robote 14", bg = "white",relief = GROOVE, wrap = WORD)
text1.place(x = 0,y = 0, width = 330,height = 123)
```

```python
scrollbar1 =  Scrollbar(f)
scrollbar1.pack(side = "right", fill = "y")
scrollbar1.configure(command = text1.yview)
text1.configure(yscrollcommand = scrollbar1.set)
```

```python
combo2 = ttk.Combobox(root,values = languageV,font = "Roboto 14", state = "r")
combo2.place(x = 53, y = 329)
combo2.set("....LANGUAGE....")
```

```python
label2 = Label(root,text = "ENGLISH",font = "segoe 30 bold", bg = "white",width = 12, bd=5, relief = GROOVE)
label2.place(x = 25, y = 365)
```

```python
f1 = Frame(root, bg = "Black",bd = 1)
f1.place(x = 10,y = 430,width = 330, height = 125)
```

```python
text2 = Text(f1, font = "Robote 14", bg = "white",relief = GROOVE, wrap = WORD)
text2.place(x = 0,y = 0, width = 330,height = 123)
```

```python
scrollbar2 =  Scrollbar(f1)
scrollbar2.pack(side = "right", fill = "y")
scrollbar2.configure(command = text2.yview)
text2.configure(yscrollcommand = scrollbar2.set)
```

```python
translate = Button(root,text = "Translate", font = ("Roboto",15), activebackground = "white",cursor = "hand2", bd = 1,
                  bg = "black", fg="white",width = 10, height=2,command = translate_now)
translate.place(x = 115, y = 255)
```

```python
label_change()
```

```python
root.mainloop()
```
